# Cases Summary by Date (Elementary)

```python
pd.read_sql_query('''
SELECT Datestamp,
       SUM(Active_Student) AS "Active Student Cases",
       SUM(Total_Student) AS "Total Student Cases",
       SUM(Active_Staff) AS "Active Staff Cases",
       SUM(Total_Staff) AS "Total Staff Cases"
FROM (  
  SELECT * FROM cases
  INNER JOIN (
    SELECT * FROM school_level WHERE Level='Elementary'
  ) USING(School)
) GROUP BY Datestamp;
''', db_conn, index_col='Datestamp', parse_dates='Datestamp')
```

# Cases Summary by Date (High)

```python
pd.read_sql_query('''
SELECT Datestamp,
       SUM(Active_Student) AS "Active Student Cases",
       SUM(Total_Student) AS "Total Student Cases",
       SUM(Active_Staff) AS "Active Staff Cases",
       SUM(Total_Staff) AS "Total Staff Cases"
FROM (  
  SELECT * FROM cases
  INNER JOIN (
    SELECT * FROM school_level WHERE Level='High'
  ) USING(School)
) GROUP BY Datestamp;
''', db_conn, index_col='Datestamp', parse_dates='Datestamp')
```

```python
df_mapplot = pd.read_sql_query('''
    SELECT
        Datestamp,
        cases.School,
        Active_Student,
        Total_Student,
        school_level.Students as "Student_Pop",
        Latitude,
        Longitude
    FROM
        cases
    LEFT JOIN school_level USING(School)
    WHERE Datestamp='{}';
    '''.format(date(2021,10,4).isoformat()), db_conn, parse_dates='Datestamp'
)
pd.set_option("display.max_rows", None, "display.max_columns", None)
df_mapplot["size"] = df_mapplot.Active_Student / df_mapplot.Student_Pop * 500
df_mapplot.dropna().plot_bokeh.map(
    x="Longitude",
    y="Latitude",
    colormap='Viridis',
    colormap_uselog=True,
    hovertool_string='''<h2> @{School} </h2>
                        <h3> Active Student Cases: @{Active_Student} </h3>
                        <h3> Total Student Cases: @{Total_Student} </h3>''',
    size="size"
)
```
