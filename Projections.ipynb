```python
import pandas as pd
import pandas_bokeh
pandas_bokeh.output_notebook()
pd.set_option('plotting.backend', 'pandas_bokeh')
import numpy as np
import sqlite3
import matplotlib
from datetime import date, timedelta
db_conn = sqlite3.connect('cases.db')
total_students = 85000
%load_ext autoreload
%autoreload 2
import techniques
```

```python
df_summary = techniques.summarize(db_conn, total_students)

df_proj_high = techniques.model(
    df_summary,
    total_students,
    seroprevalence = 0,
#    r0 = 1.62,
    r0 = 1.42,
    quarantine_factor = 7.5,
    pct_long_covid = 0.135,
    pct_severe = 0.075,
    pct_death = 0.0001,
    quarantine_period = 10
)

df_proj_low = techniques.model(
    df_summary,
    total_students,
    seroprevalence = 0.25,
#    r0 = 1.62,
    r0 = 1.42,
    quarantine_factor = 7.5,
    pct_long_covid = 0.135,
    pct_severe = 0.075,
    pct_death = 0.0001,
    quarantine_period = 10
)

df_proj_high
```

```python
df = pd.DataFrame({
    'Actual Active Cases' : df_summary['Active Student Cases'],
    'Projected Active Cases (High)' : df_proj_high['Active Cases'][:4],
    'Projected Active Cases (Low)'  : df_proj_low['Active Cases'][:4]
})
df.interpolate(method='time', limit_area='inside').plot()
```

```python
df = pd.DataFrame({
    'Actual Quarantine' : df_summary['Quarantined Students'],
    'Projected Quarantine (High)' : df_proj_high['Quarantined'][:4],
    'Projected Quarantine (Low)'  : df_proj_low['Quarantined'][:4]
})
df.interpolate(method='time', limit_area='inside').plot()
```
